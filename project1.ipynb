```python
# import needed libraries
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
# initialize the number of valus
N = 10000

# initialize values of sx and sy
sx= np.zeros(N)
sy = np.zeros(N)

# update the values of sx and sy
for i in range(1,N):
    # k is given as been any random integer between 1 and 3 inclusive
    k = randint(1,3)
    sx[i] = sx[i-1]/2 + k-1
    sy[i] = sy[i-1]/2
    
    if k ==2:
        sy[i] += 2
        

plt.plot(sx,sy,".",markersize=1)
plt.show()
```
